```python
import datetime
import requests
import numpy as np
import pandas as pd
import pprint
import math
import json
```

```python
# function to use requests.post to make an API call to the subgraph url
def run_query(q):

    # endpoint where you are making the request
    request = requests.post('https://api.thegraph.com/subgraphs/name/messari/uniswap-v3-arbitrum'
                            '',
                            json={'query': query})
    if request.status_code == 200:
        return request.json()
    else:
        raise Exception('Query failed. return code is {}.      {}'.format(request.status_code, query))

query = """

{
  liquidityPool(id: "0xc6f780497a95e246eb9449f5e4770916dcd6396a") {
    dailySnapshots(orderBy: timestamp, orderDirection: desc) {
      dailyVolumesUSD
      totalValueLockedUSD
      timestamp
    }
  }
}
"""
result1 = run_query(query)

```

```python
df = pd.DataFrame(result1['data']['liquidityPool']['dailySnapshots'])
df['datetime'] = df['timestamp'].apply(lambda x: datetime.datetime.utcfromtimestamp(int(x)))
display(df)
```
