```python
import datetime

time_from = 1680508800
time_to = 1687708800

# Convert Unix timestamps to datetime objects
datetime_from = datetime.datetime.fromtimestamp(time_from)
datetime_to = datetime.datetime.fromtimestamp(time_to)

# Convert datetime objects to formatted strings
formatted_from = datetime_from.strftime("%Y-%m-%d %H:%M:%S")
formatted_to = datetime_to.strftime("%Y-%m-%d %H:%M:%S")

print("Time From:", formatted_from)
print("Time To:", formatted_to)
```
